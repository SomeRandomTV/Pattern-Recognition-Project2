```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_excel("./Proj2&3DataSet.xlsx")

```

```python
print("=========== Details ===========")
df.columns = ["x_1", "x_2", "class"]
print(df.head())
print(df.info())

print("========== Classes ===========")
c_1 = df[df["class"] == -1]
c_2 = df[df["class"] == 1]
print(f"C_1 has {len(c_1)} elements")
print(f"C_2 has {len(c_2)} elements")

X = df[["x_1", "x_2"]]
Y = df["class"]
```

```python
plt.scatter(x=X["x_1"], y=X["x_2"], c=Y)
plt.xlabel("x_1")
plt.ylabel("x_2")
plt.show()
```
